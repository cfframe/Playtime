Compare methods for counting how many of each value in a list.

`Counter` is much faster than using a list comprehension (by an order of 300 with a list of 2000 integers)

```python
from collections import Counter
```

```python
l = [x for x in range(1000)]
l = l + l
print(f'length l: {len(l)}')
```

```python
%%timeit
Counter(l)
```

```python
Counter(l)
```

```python
%%timeit
[[x,l.count(x)] for x in set(l)]
```
